```python
from random import normalvariate as nv
```

```python
for x in range (10):
    val1 = int(nv(350,50))
    val2 = int(nv(180,30))
    print(val1, " ", val2)
```

```python
n=1000
values=[]
frequency={}

while len(values) < n:
    value = int(nv(350,50))
    if 280 < value < 420:
        frequency[value] = frequency.get(value, 0) + 1 # get function fetches the value in the dictionar & 0 is set as default.
        values.append(value)
```

```python
print(frequency)
```

```python
freq = list(frequency.items())
freq.sort()
print(freq)
```

```python
from matplotlib import pyplot as plt
plt.plot(*list(zip(*freq)))
plt.show()
```
