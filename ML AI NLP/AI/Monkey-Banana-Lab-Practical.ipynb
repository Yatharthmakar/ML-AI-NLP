```python
import numpy as np
```

```python
class room:
    def __init__(self):
        self.room_size = np.array([4,4])
        self.room= np.zeros((5,5), dtype=int)
        self.banana_pos = np.array([2,2])
        self.monkey_pos = np.array([0,0])
        self.box_pos = np.array([4,4])
        self.jump_height = 1
        self.carry_box = 0
        return
    
    def upadte_objects(self):
        self.room = np.zeros((5,5), dtype=int)  
        if self.carry_box:
            self.room[tuple(self.banana_pos)] = 3
            self.room[tuple(self.monkey_pos)] = 1
        else:
            self.room[tuple(self.banana_pos)] = 3
            self.room[tuple(self.monkey_pos)] = 1
            self.room[tuple(self.box_pos)] = 2
        return
    
    def move_right(self, moves):
        if (self.monkey_pos[1]+moves) > self.room_size[1]:
            print("Can Move this far!!")
        else:
            self.monkey_pos[1] += moves
            mon.upadte_objects()
            print("Moveed right by:", moves)
        return
    
    def move_left(self, moves):
        if (self.monkey_pos[1]-moves) < 0:
            print("Can Move this far!!")
        else:
            self.monkey_pos[1] -= moves
            mon.upadte_objects()
        return
    
    def move_down(self, moves):
        if (self.monkey_pos[0]+moves) > self.room_size[0]:
            print("Can Move this far!!")
        else:
            self.monkey_pos[0] += moves
            mon.upadte_objects()
        return
    
    def move_up(self, moves):
        if (self.monkey_pos[0]-moves) <0:
            print("Can Move this far!!")
        else:
            self.monkey_pos[0] -= moves
            mon.upadte_objects()
        return
    
    def carry(self):
        self.jump_height = 3
        self.carry_box = 1
        mon.upadte_objects()
        return
    
    def jump(self):
        print("jumped")
        if np.array_equal(self.monkey_pos, self.banana_pos) and self.jump_height == 3:
            print("Got Bananas")
            return 1
        else:
            return 0
```
